**Importing necessary Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Reading the Dataset**

```python
data = pd.read_csv("weatherHistory.csv")
```

```python
data.head()
```

**Cleaning and Processing the Dataset**

```python
data['Temperature (c)'] = data['Temperature (C)'].astype('float')

data['Apparent Temperature (C)'] = data['Apparent Temperature (C)'].astype('float')

data['Humidity'] = data['Humidity'].astype('float')

data['Wind Speed (km/h)'] = data['Wind Speed (km/h)'].astype('float')

data['Wind Bearing (degrees)'] = data['Wind Bearing (degrees)'].astype('float')

data['Visibility (km)'] = data['Visibility (km)'].astype('float')

data['Loud Cover'] = data['Loud Cover'].astype('float') 

data['Pressure (millibars)'] = data[ 'Pressure (millibars)'].astype('float')
```

```python
data['Precip Type'].fillna(data['Precip Type'].value_counts().index[0], inplace=True)
```

```python
data.isnull().sum()
```

```python
sns.heatmap(data.isnull(), yticklabels=False, cbar=False)
```

**Performing Exploratory Data Analysis**

Analyzing the most frequent weather from the ‘Summary’ column

```python
Weather= data['Summary'].value_counts().reset_index()
Weather.columns=['weather', 'Count']
Weather
```

A line plot representing the above frequencies is shown below. We can observe that the most common weather that prevails is ‘Partly Cloudy’

```python
sns.set(rc={'figure.figsize': (8,4)})
plt.xticks(rotation=90)
sns.lineplot(x=Weather ['weather'], y=Weather ['Count'], data=Weather)
plt.show()
```

Weather v/s Temperature

```python
plt.figure(figsize=(12,6))
plt.xticks (rotation = 90)
plt.title('Weather v/s Temperature')
sns.barplot (x= data['Summary'],y=data['Temperature (C)'])
```

We can observe from the above graph that the maximum temperature is when the weather is ‘Dry’.

Weather v/s Humidity

```python
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
plt.title('Weather v/s Humidity')
sns.barplot (x= data['Summary'],y=data['Humidity'])
```

The above graph remarks that the maximum humidity is for the weather types — Foggy, Breezy and Foggy, and Rain.

Weather v/s Pressure

```python
plt.figure( figsize=(8,4))
plt.xticks (rotation=90)
plt.title('Weather v/s Pressure')
sns.lineplot(y=data['Pressure (millibars)'], x=data['Summary'])
```
